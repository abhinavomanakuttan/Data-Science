````python
# 1
def evens_and_odds(n):
    count_evens = 0
    count_odds = 0
    for i in range(n+1):
        if i % 2 == 0:  # To check even 
            count_evens += 1 #Output: 51
        else:
            count_odds += 1 #Output: 50
        
    print(f"Count of even numbers {count_evens} \nCount of odd numbers {count_odds}")
             
evens_and_odds(100)
````

````python
def factorial(n):
    # Check if the input is a non-negative integer
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1:
        return 1
    
    result = 1
    # Loop through the range from 1 to n and multiply
    for i in range(1, n + 1):
        result *= i
    
    return result

# Example usage
print(factorial(5))  # Output: 120
print(factorial(0))  # Output: 1
print(factorial(-3)) # Output: Factorial is not defined for negative numbers.
````

````python
# 3
def _is_empty_(parameter):
    # Check if the parameter is empty
    if not parameter:
        return True
    else:
        return False

# Example usage
print(_is_empty_([]))        # Output: True (empty list)
print(_is_empty_({}))        # Output: True (empty dictionary)
print(_is_empty_(''))        # Output: True (empty string)
print(_is_empty_([1, 2, 3])) # Output: False (list is not empty)
print(_is_empty_('Hello'))   # Output: False (string is not empty)
````

````python
# 4
from statistics import mode
import math

# Function to calculate the mean
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Function to calculate the median
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2

    if n % 2 == 0:  # even number of elements
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:  # odd number of elements
        return sorted_numbers[mid]

# Function to calculate the mode
def calculate_mode(numbers):
    try:
        return mode(numbers)
    except:
        return "No unique mode found"  # In case there's no unique mode

# Function to calculate the range
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# Function to calculate the variance
def calculate_variance(numbers):
    mean = calculate_mean(numbers)
    squared_diffs = [(x - mean) ** 2 for x in numbers]
    return sum(squared_diffs) / len(numbers)

# Function to calculate the standard deviation
def calculate_std(numbers):
    variance = calculate_variance(numbers)
    return math.sqrt(variance)

# Example usage
data = [10, 20, 20, 30, 40, 50, 60]

print("Mean:", calculate_mean(data))           # Output: 32.857
print("Median:", calculate_median(data))       # Output: 30
print("Mode:", calculate_mode(data))           # Output: 20
print("Range:", calculate_range(data))         # Output: 50
print("Variance:", calculate_variance(data))   # Output: 276.53
print("Standard Deviation:", calculate_std(data)) # Output: 16.63
````
